```python
import os
import random
import shutil #thư viện để sao chép-di chuyển file
```

```python
#đường dẫn tới thư mục gốc
origin_dir='/Users/namtran/Downloads/cats and dogs'
origin_cats_dir=os.path.join(origin_dir,'cat')
origin_dogs_dir=os.path.join(origin_dir,'dog')
                        
#đường dẫn tới thư mục trong project
new_dir='./data'
train_dir=os.path.join(new_dir,'train')
validation_dir=os.path.join(new_dir,'validation')

train_cats_dir=os.path.join(train_dir,'cats')
train_dogs_dir=os.path.join(train_dir,'dogs')
validation_cats_dir=os.path.join(validation_dir,'cats')
validation_dogs_dir=os.path.join(validation_dir,'dogs')
```

```python
#tỉ lệ phân chia
split_ratio=0.2
```

```python
#tạo các thư mục nếu chưa tồn tại
os.makedirs(train_cats_dir,exist_ok=True)
os.makedirs(train_dogs_dir,exist_ok=True)
os.makedirs(validation_cats_dir,exist_ok=True)
os.makedirs(validation_dogs_dir,exist_ok=True)
print(f"Đã tạo các thứ mục {train_cats_dir}, {train_dogs_dir}, {validation_cats_dir}, {validation_dogs_dir} ")
```

```python
#lấy danh sách ảnh từ thư mục gốc
try:
    cat_filenames=[fname for fname in os.listdir(origin_cats_dir) if fname.endswith ('.jpg')]
    dog_filenames=[fname for fname in os.listdir(origin_dogs_dir) if fname.endswith ('.jpg')]
    
except FileNotFoundERROR:
    print(f"end=point {origin_cats_dir} or {origin_dogs_dir} not found")
    exit()
```

```python
#xáo trộn và tính toán số lượng ảnh
random.shuffle(cat_filenames)
random.shuffle(dog_filenames)
print(f"Tìm thấy {len(cat_filenames)} ảnh mèo đã xáo trộn trong {origin_cats_dir}")
print(f"Tìm thấy {len(dog_filenames)} ảnh chó đã xáo trộn trong {origin_dogs_dir}")
```

```python
#tính toán số lượng ảnh cho mỗi tập 
num_cats_validation=round(len(cat_filenames)*split_ratio)
num_dogs_validation=round(len(dog_filenames)*split_ratio)
num_cats_train=len(cat_filenames)-num_cats_validation
num_dogs_train=len(dog_filenames)-num_dogs_validation
print(f" {num_cats_train} ảnh mèo và {num_dogs_train} ảnh chó trong tập train")
print(f" {num_cats_validation} ảnh mèo và {num_dogs_validation} ảnh chó trong tập validation")
```

```python
#hàm sao chép
def copied_files(filenames,src_dir,destination_dir):
    copied_count=0
    for fname in filenames:
        src_path=os.path.join(src_dir,fname)
        dst_path=os.path.join(destination_dir,fname)
        try:
            shutil.copyfile(src_path,dst_path)
            copied_count+=1
        except Exception as e:
            print(f"Không thể sao chép {src_path} đến {dst_path}: {e}")
        print(f"Đã sao chép {copied_count} ảnh từ {src_dir} đến {destination_dir}")
    return copied_count        
        
```

```python
#sao chép ảnh mèo
print("Sao chép ảnh mèo vào thư mục train và validation")
copied_cats_train=copied_files(cat_filenames[:num_cats_train],origin_cats_dir,train_cats_dir)
copied_cats_validation=copied_files(cat_filenames[:num_cats_validation],origin_cats_dir,validation_cats_dir)
print(f"Đã hoàn thành sao chép ảnh")
```

```python
#sao chép ảnh chó
print(f"Sao chép ảnh chó vào thư mục train và validation")
copied_dogs_train=copied_files(dog_filenames[:num_dogs_train],origin_dogs_dir,train_dogs_dir)
copied_dogs_validation=copied_files(dog_filenames[:num_dogs_validation],origin_dogs_dir,validation_dogs_dir)
print(f"Đã hoàn thành sao chép ảnh")
```
